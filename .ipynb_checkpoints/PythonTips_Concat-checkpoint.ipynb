## Libraries

```python
# Importing pandas with an alias
import pandas as pd
```

## Creating DataFrames

```python
# Simple creation of DataFrame1
df1 = pd.DataFrame({"name": ["Caio", "Rodrigo", "Evandro", "Rafael"],
                     "age": [28, 45, 27, 16]})
df1
```

```python
# Simple creation of a DataFrame2
df2 = pd.DataFrame({"name": ["Caio", "Mariane", "Maria"],
                     "age": [28, 29, 59]})
df2
```

## CONCAT simulating SQL UNION ALL

```python
# Using concat() function to union DataFrame1 and DataFrame2, ignoring their indexes
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
# Using reset_index() function to ignore their indexes (option without ignore_index in concat())
df.reset_index(drop=True, inplace=True)
df
```

## CONCAT simulating SQL UNION

```python
# Using concat() function to union DataFrame1 and DataFrame2, using drop_duplicates() function to simulate SQL UNION)
DF = pd.concat([df1, df2]).drop_duplicates().reset_index(drop=True)
DF
```

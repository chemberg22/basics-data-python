### Dictionaries definition: is an object represented by a collection of 1 or more objects. Each object has your key with your value. To access the value of an object we just need to call your key. These objects are started and ended by braces {'x' : 'y'}

#### Creating dictionaries

```python
# Simple statement
dict1 = {'key1': 'value1', 
         'key2': 'value2'}
print(dict1)
type(dict1)
```

```python
# Using dict() simple function
dict2 = dict(a = 1, 
             b = 2, 
             c = 3, 
             d = 'a')
print(dict2)
```

```python
# Using dict() function, receiving a list of tuples
dict3 = dict([('a', 1), 
              ('b', 2), 
              ('c', 3)])
print(dict3)
```

```python
# Simple statement with key > list of values
dict4 = {'key1': ['value1', 2, 3], 
         'key2': [2, 'age', 'name']}
print(dict4)
```

```python
# Simple statement to create a dictionary of a dictionary
dict5 = {'id': [1, 2, 3], 
         'user': {'name': ['Matheus', 'Rodrigo', 'Rafael'], 
                  'age': [25, 30, 29]}}
print(dict5)
```

```python
# Using zip() function
dict6 = dict(zip(['key1', 'key2', 'key3'], 
                 ['value1', 'value2', 'value3']))
print(dict6)
```

```python
# Simple statement to create a dictionary of a tuple
dict7 = {'names': ('Matheus', 'Rafael'), 
         'ages': (25, 30)}
print(dict7)
```

### Acessing dictionary values

```python
# Using the needed key name
print(dict1)
dict1['key1']
```

```python
# Using get() function with key name
print(dict1)
dict1.get('key1')
```

```python
# Using get() function with a key name, if the key doesn't exist, it will return a default value
print(dict1)
dict1.get('key5', 'Non-existent key!')
```

```python
# Using get() function with list() function to return the value of the indexed key
print(dict1)
dict1.get(list(dict1)[1])
```

```python
# Using FOR loop with keys() function to iterating through the dictionary keys
print(dict1)
for key in dict1.keys():
    print(key, dict1[key])
```

```python
# Using FOR with items() function to iterating through the dictionary keys and values
print(dict1)
for key, value in dict1.items():
    print(key, value)
```

### Gathering dictionaries

```python
# Using update() function to gather dictionaries
print(dict1)
print(dict2)
dict1.update(dict2)
print(dict1)
```

```python
# Adding a new key to the dictionary
print(dict1)
dict1["key"] = 1000
print(dict1)
```

```python
# Using update() function to add a new key to the dictionary
print(dict2)
dict2.update({'key': 1000})
print(dict2)
```

```python
# Using update() function to update a key of the dictionary
print(dict3)
dict3.update({'c': 5})
print(dict3)
```

```python
# Gathering dictionaries and updating equal keys
print(dict2)
print(dict3)
{**dict2, **dict3}
```

### Verifyng if a value belongs to a key

```python
# Using equal comparison
print(dict3)
1 == dict3['a']
```

```python
# Using equal comparison
print(dict3)
0 == dict3['a']
```

### Methods

```python
# Using clear() function to clear a dictionary
print(dict3)
dict3.clear()
print(dict3)
```

```python
# Using copy() function to clone a dictionary in a new one
dict1000 = dict2.copy()
print(dict1000)
```

```python
# Using pop() function to remove a key : value from a dictionary
print(dict4)
dict4.pop('key1')
print(dict4)
```

```python
# Using popitem() function to remove the last key : value from a dictionary
print(dict6)
dict6.popitem()
print(dict6)
```

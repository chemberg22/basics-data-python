## Libraries

```python
# Importing pandas and pandasql with aliases
import pandas as pd
from pandasql import sqldf
```

## Loading data

```python
# Loading categories.csv file
categories = pd.read_csv('../../data/categories.csv')
categories
```

```python
# Loading products.csv file
products = pd.read_csv('../../data/products.csv')
products
```

## Combining information from datafames using SQL

```python
# Creating a SQL query to gather our dataframes and using sqldf() function to execute it
query = """

    SELECT 
        p.id as product_id,
        p.name as product_name,
        p.price,
        c.name as category_name
    FROM products p
    JOIN categories c on c.id = p.category_id
    
    """

result = sqldf(query)
result
```
